```python
import matplotlib.pyplot as plt
import numpy as np
fig,ax= plt.subplots()
x= np.linspace(0, 20, 1000)      #linspace -> 1000 data equally_sapced between 0 to 20 number.
ax.plot(x, np.cos(x))
ax.axis('equal')
ax.annotate('local maximum', xy=(6.26, 1), xytext=(10,4), arrowprops=dict(facecolor='black', shrink=0.06))
ax.annotate('local minimum', xy=(5*np.pi, -1), xytext=(2,-6), arrowprops=dict(arrowstyle='->', connectionstyle="angle3, angleA=0,angleB=-90"));
print(x)
```

```python
import matplotlib.pyplot as plt
w=4
h=3
d=70
plt.figure(figsize=(w,h),dpi=d)
plt.axis([0, 5, 0, 5])
x=[0, 3, 5]
y=[1, 4, 3.5]
label_x=1
label_y=4
arrow_x=3
arrow_y=4
arrow_properties=dict(facecolor='black',width=0.5, headwidth=4 ,shrink=0.1)
plt.annotate('maximum', xy=(arrow_x, arrow_y), xytext=(label_x,label_y), arrowprops=arrow_properties)
plt.plot(x,y)
```

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
fig= plt.figure()
ax=plt.axes(projection='3d')
z=np.linspace(0,1,100)
print(z)
x=z*np.sin(25*z)
y=z*np.cos(25*z)
ax.plot3D(x, y,z,'green')
ax.set_title("3D line plot")
plt.show()
```
